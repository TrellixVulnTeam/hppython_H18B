```python
import numpy as np
import pandas as pd
from sklearn.datasets import twenty_newsgroups
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
import seaborn as sns
```

```python
ng20 = twenty_newsgroups.fetch_20newsgroups(subset='train', categories=['rec.sport.baseball', 'comp.graphics', 'comp.windows.x'], remove=['headers','footers','quotes'])
from sklearn.feature_extraction.text import CountVectorizer
count_vect = CountVectorizer(stop_words='english', max_df=0.95, max_features=1000)
ng20_occur = count_vect.fit_transform(ng20.data)
```

```python
count_vect.get_feature_names()
```

```python
ng20_occur[ng20_occur>1] = 1
```

```python
ng20_occur = ng20_occur.toarray()
```

```python
ng20_occur_df = pd.DataFrame(ng20_occur, columns=count_vect.get_feature_names())
```

```python
ng20_occur_df.head()
```

```python
ng20_freq_item = apriori(ng20_occur_df, min_support=0.02, use_colnames=True)
ng20_freq_item
```

```python
ng20_assoc_rules = association_rules(ng20_freq_item, metric="confidence", min_threshold=0.6)
ng20_assoc_rules
```
