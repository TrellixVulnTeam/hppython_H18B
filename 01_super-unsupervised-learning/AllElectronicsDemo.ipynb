```python
%matplotlib inline
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
allelectronics = pd.read_csv("AllElectronics.csv")
allelectronics = allelectronics.drop(['ID'], axis=1)
allelectronics
```

```python
allelectronics_onehot = pd.get_dummies(allelectronics, drop_first=False)
```

```python
allelectronics_freq_item = apriori(allelectronics_onehot, min_support=0.2, use_colnames=True)
allelectronics_freq_item
```

```python
allelectronics_assoc_rules = association_rules(allelectronics_freq_item, metric="confidence", min_threshold=0.8)
allelectronics_assoc_rules
```

```python
allelectronics_assoc_rules[(allelectronics_assoc_rules['support']>0.23) & (allelectronics_assoc_rules['confidence']>0.9)]
```
